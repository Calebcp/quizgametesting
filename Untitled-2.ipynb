## 1. Pointer Concept
### key points
-memory model in C (how variables stored in memory, addresses)
-Pointer definition: **A pointer is avariable that stores the address of another variable.**
-the addresss is meaningless to us, but if computer knows the address it can use the address to acces whatever is in the address.
-syntax: int *p;
- & operator (get address) and * operator (dereference)
-NULL pointer
-Danger of unitialized/dangling pointers
-printing addresses with %p
-**pointer size is usually the same regardless ofo type.**
beause the addresss space of computer is fixed, and any pointer is just an address, the type of pointer is type of address you set in in it

```python
%%file pointer.c
#include <stdio.h>
int main() {
    int a = 10;
    int *p = &a; // pointer to a
    printf("Value of a: %d\n", *p); // dereferencing pointer
    printf("Address of a: %p\n", (void*)&a);
    printf("Address stored in p: %p\n", (void*)p);
    *p = 20; // modifying value of a through pointer
    printf("New value of a: %d\n", a);
    return 0;
```
