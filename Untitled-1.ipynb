```python
%% file binarySearch.c
#include <stdio.h>
int main()
{
    int numbers[]={10,20,25,30,40,50}; //initial sorted array
    int n=6; // number of elements
    int target=25; // element to search
    int left=0, right=n-1;
    int foundIndex=-1; //index of found element
    while(left<=right)
    {
        int mid=(left+right)/2;
        if(numbers[mid]==target)
        {
            foundIndex=mid;
            break;
        }
        else if(numbers[mid]<target)
        {
            left=mid+1;
        }
        else
        {
            right=mid-1;
        }
    }
    if(foundIndex!=-1)
    {
        printf("Element %d found at index %d\n", target, foundIndex);
    }
    else
    {
        printf("Element %d not found in the array\n", target);
    }
    return 0;
}
```

```python
##file bubbleSort.c
#include <stdio.h>
int main()
{
    int numbers[]={30,10,50,20,40}; //initial unsorted array
    int n=5; // number of elements
    for(int i=0; i<n-1; i++) //number of passes
    {
        for(int j=0; j<n-i-1; j++) //compare adjacent elements
        {
            if(numbers[j]>numbers[j+1])
            {
                //swap
                int temp=numbers[j];
                numbers[j]=numbers[j+1];
                numbers[j+1]=temp;
            }
        }
    }
    printf("Sorted array:\n");
    for(int i=0; i<n; i++)
    {
        printf("%d ", numbers[i]);
    }
    printf("\n");
    return 0;
}

//to solve something like this , to draw it and how it moves, is very helpful
//if you just look at the code and try to imagine it, it is very difficult
//after-class exercise: implement selectionSort and insertionSort functions..-
//sorting algorithms with c langauge youtube videos
```
